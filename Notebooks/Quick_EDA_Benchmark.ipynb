## Model

For the model we have to make feature engineering, this will be changing the categorical (object) columns into one hote encoding or label encoding, this will be just checking how it will work for the model. For now, the correlation heat map is only showing correlation with the numerical columns, and we need to analyze if the other columns have something to do with our model.

We can model right now and then make the feature engineering. Will be creating a base line model.

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression

numeric_specific_df = df.select_dtypes(include=['int64', 'float64'])
numeric_specific_df
```

```python
X = numeric_specific_df.drop(columns={'fraudulent'})
Y = numeric_specific_df[['fraudulent']].values

X_train, X_test, Y_train, Y_test = train_test_split(X, Y)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y)

LgR = LogisticRegression()
LgR.fit(X_train, Y_train)
Y_pred = LgR.predict(X_test)
Y_test = Y_test.flatten()
Y_pred = Y_pred.flatten()
```

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
accuracy_score(Y_pred, Y_test)
```

```python
cm = confusion_matrix(Y_test, Y_pred, labels=LgR.classes_)
disp = ConfusionMatrixDisplay(confusion_matrix=cm,
                               display_labels=LgR.classes_)
disp.plot()
plt.show()
```

95% of accuracy is bad, and that is because we know that our target is ovesampled with real jobs. So if you see the confusion matrix it is only predicting real no fake jobs, which is bad. So here it comes the feature engineering
